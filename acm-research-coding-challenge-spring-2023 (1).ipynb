**The Color indicates the Temperature**
* The Colder the temperature of the Star, the more of a reddish color it will be (red, orange). As the stars get warmer, specifically around the 5000 Kelvin(K) mark, the stars become more white. Finally when the stars get the hottest they can be, aorund the 7000-8000K mark, they become more blue.

**The Temperature can also indicate what type of star it is**
* The colder stars seem to be more of Brown and Red Dwarfs, wheras the Main Sequence stars are generally the warmest. However, Supergiants and Hypergiants seem to vary wildly, with some hypergiants being just as cold as the dwarfs.

**The Size of the star closely corrilates with the Luminosity**
